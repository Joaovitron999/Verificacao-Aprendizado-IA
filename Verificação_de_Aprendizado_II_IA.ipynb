Considerações sobre as colunas:

1. A coluna *`id`* representa o número de identificação único de cada instância.

2. A coluna *`runtime`* representa o número de vezes que o título foi reproduzido

3. A coluna *`imdb_score`* representa a classificação do título com base de votações dos usuários.

4. Os demais atributos *(outros 12)* são strings e medidas numéricas que caracterizam cada instância.

```
# Removendo colunas desnecessárias:
#--------------------------------------------------

series = series.drop(columns=['release_year','age_certification','genres','production_countries','imdb_id','description','seasons','tmdb_popularity','tmdb_score'])
```

```
# Removendo valores inconsistentes
#--------------------------------------------------

series = series.drop(series[np.isnan(series['runtime'])].index)
series = series.drop(series[np.isnan(series['imdb_score'])].index)
series = series.drop(series[np.isnan(series['imdb_votes'])].index)
```

```
# Prevendo valores imdb_score 
# e gerando a janela deslizante
#--------------------------------------------------

valores = list(series.imdb_score)                                                         # adicionando os valores da coluna "imdb_score" a uma lista

tamanhoJanela = 5                                                                         # tamanho da janela
linhasJanela = []                                                                         # iniciando lista de listas, onde cada elemento sera uma linha da a ser exibida pela janela deslizante
colunas = []                                                                              # lista para salvar os rotulos das colunas da janela deslizante

#gerando rotulos das colunas
for i in range(tamanhoJanela - 1):
    colunas.append("obj" + str(i + 1))

colunas.append("Valor desejado")

i = 0
while not i + tamanhoJanela > len(valores) - 1:                                           # while para que a janela deslizante não tente exibir uma linha que tenha menos valores que seu tamanho
    linhaI = []                                                                           # lista que ira armazenar a linha da janela sendo gerada nesta iteração

    for j in range(i, i + tamanhoJanela):                                                 # gerando linha da janela e salvando em uma lista
        linhaI.append(valores[j])
    
    linhasJanela.append(linhaI)                                                           # salvando a linha gerada em uma lista
    i = i + 1

linhasJanela = np.array(linhasJanela)                                                     # transformando a lista de linhas num array numpy

dataFrame = pd.DataFrame(data=linhasJanela, columns=colunas)                              # gerando Data Frame da janela deslizante

dataFrame
```

1. A coluna *`Valor desejado`* representa a saída de cada instância imdb que desliza uma janela de 5 em 5.
